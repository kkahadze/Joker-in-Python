```python
import random

class Card(object):
    def __init__(self, value, suit):
        self.value = value
        self.suit = suit
        self.showing = True

    def __repr__(self):
        value_name = ""
        suit_name = ""
        if self.showing:
            if self.value == 4:
                value_name = "Six"
            elif self.value == 5:
                value_name = "Seven"
            elif self.value == 6:
                value_name = "Eight"
            elif self.value == 7:
                value_name = "Nine"
            elif self.value == 8:
                value_name = "Ten"
            elif self.value == 9:
                value_name = "Jack"
            elif self.value == 10:
                value_name = "Queen"
            elif self.value == 11:
                value_name = "King"
            elif self.value == 12:
                value_name = "Ace"
            elif self.value == 13:
                value_name = "Joker"
            if self.suit == 0:
                suit_name = "Diamonds"
            elif self.suit == 1:
                suit_name = "Clubs"
            elif self.suit == 2:
                suit_name = "Hearts"
            elif self.suit == 3:
                suit_name = "Spades"
            if value_name == "Joker":
                return "Joker"
            else:
                return value_name + " of " + suit_name
        else:
            return "[CARD]"

class StandardJokerDeck(list):
    def __init__(self):
        super().__init__()
        suits = list(range(4))
        values = list(range(5, 13))
        # Ranks 7 through A are added
        [[self.append(Card(i, j)) for j in suits] for i in values]
        # Ranks Six and Ace are added
        self.extend([Card(13, 0), Card(13, 0), Card(6, 0), Card(13, 0)]) 

    def __repr__(self):
        return f"Standard deck of cards\n{len(self)} cards remaining"

    def shuffle(self):
        random.shuffle(self)
        print("\n\n--deck shuffled--")

    def deal(self, location, times=1):
        for i in range(times):
            location.cards.append(self.pop(0))

    def burn(self):
        self.pop(0)

    def playable(self, wildsuit, first_suit):
        first_and_joks = filter(lambda x: x.suit == first_suit)
        if first_and_joks == []:
            filter()
        

class Player(object):
    def __init__(self, id_in):
        self.id = id_in
        self.score = 0
        self.cards = []
        self.called = 0
        self.taken = 0

    def __repr__(self):
        id = self.id
        return id

class Game(object):
    def __init__(self):
        self.deck = StandardJokerDeck()
        self.wildcard = None
        self.first_suit = None
        self.cards_dealt = 0
        self.round = None
        self.play = 0
        self.users = []
        self.dealer = None

    def get_place_in_playing_order(self, player_id):
        dealer = self.dealer
        for i in range(1,4):
            if (self.dealer + i) % 4 == player_id:
                out = i
        out

    def get_wildsuit(self):
        if self.wildcard.value == 13:
            return 4
        elif self.wildcard.suit == 0:
            return 0
        elif self.wildcard.suit == 1:
            return 1
        elif self.wildcard.suit == 2:
            return 2
        elif self.wildcard.suit == 3:
            return 3

    def next_dealer(self):
        (self.dealer + 1) % 4

    def set_dealer(self):
        self.dealer = self.next_dealer()
```
